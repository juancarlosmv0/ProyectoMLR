# Procesar datos

```r
# Leyendo los datos crudos, provenientes de la estacion Ensenada, Baja California
# Extraidos de 76050099999 en https://www.ncei.noaa.gov/access/search/data-search/global-hourly?pageNum=1&stations=76050099999
datos02 <- read.csv("2002.csv")
summary(datos02)
datos01 <- read.csv("2001.csv")
summary(datos01)
datos00 <- read.csv("2000.csv")
summary(datos00)
datos99 <- read.csv("1999.csv")
summary(datos99)
datos98 <- read.csv("1998.csv")
summary(datos98)
# Estos datasets contienen muchos datos de lecturas en crudo, datos faltantes erroneas etc,
# ademas contiene varias columnas con nombre codificado para las cuales no encontramos interpretacion.
```

```r
# Extrayendo las columnas relevantes para este caso.
columnas <- c("DATE","WND","VIS","TMP","DEW","SLP")
datos02 <- datos02[columnas]
datos01 <- datos01[columnas]
datos00 <- datos00[columnas]
datos99 <- datos99[columnas]
datos98 <- datos98[columnas]
# Concatenaremos los datasets de modo que tengamos un dataset de 5 anios cconsecutivos
datosEns <- rbind(datos98,datos99,datos00,datos01,datos02)
datosEns
```

```r
# Instalando librerias necesarias
install.packages("dplyr")
library(dplyr)
install.packages("nlme")
library(nlme)
```

```r
# Podemos ver que las observaciones (datos de cada celda) estan capturadas como caracteres y contienen codigos de medicion (ya sea de calidad o presicion, etc).
# Debemos procesar, limpiar estos datos y extraer las magnitudes adecuadas.
# Despues de investigar las unidades de medida estandares utilizadas en datos atmosfericos realizamos los siguiente:
# Funciones definidas para limpiar y convertir los datos.

# La columna TMP la temperatura esta expresada en los primeros caracteres como +0000 en centigrados (+000.0).
formatotemp <- function(temp) {
  # Extraer los caracteres necesarios (ninguna temperatura llega a negativo asi que ignoramos el +).
  temperatura <- as.numeric(substr(temp, 1, 5))/10
  return(temperatura)}

# En la columna SLP la presion al nivel del mar en hPa (0000.0).
formatopress <- function(slp) {
  press <- as.numeric(substr(slp, 1, 5))/10
  return(press)}

# En la columna WND tenemos la direccion del viento en grados en los primeros caracteres y la velocidad en km/h en los caracteres 9-12
# Convertimos los grados en 8 direcciones segun los puntos cardinales.
formatodir <- function(direccion) {
  dir_labels <- c("N", "NE", "E", "SE", "S", "SW", "W", "NW")
  dir_breaks <- c(-1, 22.5, 67.5, 112.5, 157.5, 202.5, 247.5, 292.5, 337.5)
  direccion <- ifelse(direccion == 999, NA, direccion)
  # Realizamos esto para crear una columna categórica basada en los rangos
  direccion <- cut(direccion, breaks = dir_breaks, labels = dir_labels, include.lowest = TRUE)
  #direccion <- factor(direccion, levels = dir_labels, ordered = TRUE)
  return(direccion)}

# En la columna DATE tenemos fecha y hora, extraemos estos datos por separado.
# Ademas observamos que solo tenemos horas 00:00 y 18:00, creamos otra columna categorica.
formatodate <- function(cdate){
  # Convertir la columna DATE a fecha reconocible por R.
  cdate <- as.POSIXct(cdate, format = "%Y-%m-%dT%H:%M:%S")
  # Extraer la fecha, mes y la hora separadas
  datedf <- data.frame(date = as.Date(cdate),
          time = format(cdate, "%H:%M:%S"),
          time_cat = ifelse(format(cdate, "%H:%M:%S") == "00:00:00", "A",
                      ifelse(format(cdate, "%H:%M:%S") == "12:00:00", "B",
                             ifelse(format(cdate, "%H:%M:%S") == "18:00:00", "C",
                                    ifelse(format(cdate, "%H:%M:%S") == "21:00:00", "D", NA)))))
  datedf$month <- factor(month.name[as.numeric(format(cdate, "%m"))], levels = month.name)
  #datedf$time_cat <- factor(datedf$time_cat, levels = c("A", "B", "C", "D"), ordered = TRUE)
  return(datedf)}
```

```r
# Procesando los datos de los dataframes originales asignandolos a nuevos dataframes.
# Los seguimos manteniendo por separado por si acaso.
datos02a <- datos02 %>% transmute(day = formatodate(DATE)$date, # extraer la fecha
                        time = formatodate(DATE)$time, # extraer la hora
                        timecat = as.factor(formatodate(DATE)$time_cat), #A: 00:00, B: 12:00, C: 18:00, D: 21:00
                        month = formatodate(DATE)$month,
                        WNDdir = as.numeric(substr(WND, 1, 3)), #direccion del viento en grados
                        WNDpc = formatodir(WNDdir), #direccion del viento en puntos cardinales
                        WNDvel = as.numeric(substr(WND, 9, 12))/10, #velocidad del viento capturada en km/h
                        VIS = as.numeric(substr(VIS, 1, 6))/1000, #visibilidad capturada en kilometros
                        TMPc = formatotemp(TMP), #temperatura en celsius
                        DEWc = formatotemp(DEW), #punto de rocio en celsius
                        SLP = formatopress(SLP)) #presion al nivel del mar en hecto pascales
datos01a <- datos01 %>% transmute(day = formatodate(DATE)$date,
                        time = formatodate(DATE)$time,
                        timecat = as.factor(formatodate(DATE)$time_cat),
                        month = formatodate(DATE)$month,
                        WNDdir = as.numeric(substr(WND, 1, 3)),
                        WNDpc = formatodir(WNDdir),
                        WNDvel = as.numeric(substr(WND, 9, 12))/10,
                        VIS = as.numeric(substr(VIS, 1, 6))/1000,
                        TMPc = formatotemp(TMP),
                        DEWc = formatotemp(DEW),
                        SLP = formatopress(SLP))
datos00a <- datos00 %>% transmute(day = formatodate(DATE)$date,
                        time = formatodate(DATE)$time,
                        timecat = as.factor(formatodate(DATE)$time_cat),
                        month = formatodate(DATE)$month,
                        WNDdir = as.numeric(substr(WND, 1, 3)),
                        WNDpc = formatodir(WNDdir),
                        WNDvel = as.numeric(substr(WND, 9, 12))/10,
                        VIS = as.numeric(substr(VIS, 1, 6))/1000,
                        TMPc = formatotemp(TMP),
                        DEWc = formatotemp(DEW),
                        SLP = formatopress(SLP))
datos99a <- datos99 %>% transmute(day = formatodate(DATE)$date,
                        time = formatodate(DATE)$time,
                        timecat = as.factor(formatodate(DATE)$time_cat),
                        month = formatodate(DATE)$month,
                        WNDdir = as.numeric(substr(WND, 1, 3)),
                        WNDpc = formatodir(WNDdir),
                        WNDvel = as.numeric(substr(WND, 9, 12))/10,
                        VIS = as.numeric(substr(VIS, 1, 6))/1000,
                        TMPc = formatotemp(TMP),
                        DEWc = formatotemp(DEW),
                        SLP = formatopress(SLP))
datos98a <- datos98 %>% transmute(day = formatodate(DATE)$date,
                        time = formatodate(DATE)$time,
                        timecat = as.factor(formatodate(DATE)$time_cat),
                        month = formatodate(DATE)$month,
                        WNDdir = as.numeric(substr(WND, 1, 3)),
                        WNDpc = formatodir(WNDdir),
                        WNDvel = as.numeric(substr(WND, 9, 12))/10,
                        VIS = as.numeric(substr(VIS, 1, 6))/1000,
                        TMPc = formatotemp(TMP),
                        DEWc = formatotemp(DEW),
                        SLP = formatopress(SLP))
datosEnsa <- rbind(datos98a,datos99a,datos00a,datos01a,datos02a)
datosEnsa
```

```r
# Ahora podemos leer los datos con mas facilidad, notamos que tenemos mediciones del tipo 999... y datos NA, estos datos nos van a causar problemas.
# Vamos a ignorar los renglones donde tenemos datos problematicos o irrelevantes.
datosEnsb <- datosEnsa[!(datosEnsa$VIS >= 60 | datosEnsa$VIS==0 | datosEnsa$WNDvel >= 60 | is.na(datosEnsa$WNDpc)
                         | datosEnsa$TMPc >= 80 | datosEnsa$DEWc >=  80 | datosEnsa$SLP >= 1100), ]
rownames(datosEnsb) <- NULL
# Finalmente tenemos los datos limpios
datosEnsb
```

# Analizar los datos

```r
#Observamos las diferentes graficas para ver el comportamiento entre las variables
pairs(datosEnsb[,c("VIS","WNDvel","TMPc","DEWc","SLP")])
```

# Modelo lineal multiple

```r
# Nos interesa analizar los efectos de las condiciones atmosfericas sobre la visibilidad, para poder estimar la visibilidad dadas ciertas condiciones.
# Creemos que la visibilidad se puede ver afectada por la presion atmosferica, y la velocidad y direccion del viento.
modelo1a <- lm(VIS ~ WNDvel + WNDpc, data = datosEnsb)
summary(modelo1a)
print("AIC: ")
AIC(modelo1a)
# Agregamos la presion a nivel del mar
modelo1b <- lm(VIS ~ WNDvel + WNDpc + SLP, data = datosEnsb)
summary(modelo1b)
anova(modelo1b)
print("AIC: ")
AIC(modelo1b)
```

```r
# Tambien es logico que la temperatura y el punto de rocio puedan afectar la visibiidad (los datos no contienen informacion sobre la humedad)
modelo1c <- lm(VIS ~ TMPc + DEWc + SLP, data = datosEnsb)
summary(modelo1c)
anova(modelo1c)
print("AIC: ")
AIC(modelo1c)
```

# Modelo lineal con interacciones

```r
# Los modelos anteriores no nos estan representando correctamente la visibilidad
# Proponemos un modelo de regresion multiple, agregando interaccion entre temperatura y punto de rocio
modelo2a <- lm(VIS ~ DEWc + TMPc + TMPc*DEWc + SLP, data = datosEnsb)
summary(modelo2a)
anova(modelo2a)
print("AIC: ")
AIC(modelo2a)
# Aqui empezamos a observar relaciones mas significvas entre algunas variables y la visibilidad
# Vemos que la velocidad del viento pierde peso si tenemos otras variables ajustadas.
modelo2b <- lm(VIS ~ DEWc + TMPc + WNDvel*WNDpc + SLP, data = datosEnsb)
summary(modelo2b)
anova(modelo2b)
print("AIC: ")
AIC(modelo2b)
# Los p valores disminuyen, sin embargo el R^2 ajustada no mejora
```

# Modelo de efectos mixtos



```r
# Los modelos anteriores muestran leve mejora en el ajuste pero debemos restringir los valores de la temperatura a mayores a 0.
# Proponemos un modelo de efectos mixtos donde la direccion del viento y el mes aportan un efecto aleatorio a la medicion
modelo3 <- lme(VIS ~ TMPc*DEWc, random = list (month = ~1, WNDpc = ~1) , data = datosEnsb)
summary(modelo3)
anova(modelo3)
```

# Modelos generalizados

```r
# Proponemos modelos usando glm con distribucion log-normal
modelo4a <- glm(VIS ~ WNDvel + TMPc*DEWc + SLP, data = datosEnsb, family = gaussian(link = "log"))
summary(modelo4a)
plot(modelo4a)
```

```r
# Proponemos un modelo de distribucion Gamma
modelo4b <- glm(VIS ~ WNDvel + TMPc*DEWc + SLP, data = datosEnsb, family = Gamma(link = "log"))
summary(modelo4b)
plot(modelo4b)
```

```r
# Para analizar la probabilidad de baja visibilidad, definimos una columna
# Creamos una variable booleana para definir un limite de baja visibilidad (15 kilometros)
datosEnsc <- datosEnsb %>%
  mutate(low_vis = ifelse(VIS < 15, 1, 0))
datosEnsc
```

```r
# Proponemos un modelo para ver que factores aportana a una baja visibilidad.
modelo4 <- glm(low_vis ~ WNDvel + TMPc * DEWc + SLP ,
                     family = binomial(link = "logit"), data = datosEnsc)
summary(modelo4)
```
